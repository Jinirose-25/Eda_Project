## How does the shipping cost per order vary across different shipping modes and regions

```python
mean_shipping_cost = df.groupby(['Region', 'Ship_mode'])['Shipping_cost'].mean()

# Group by region and shipping mode, calculate median shipping cost
median_shipping_cost = df.groupby(['Region', 'Ship_mode'])['Shipping_cost'].median()

# Group by region and shipping mode, calculate standard deviation of shipping cost
std_shipping_cost = df.groupby(['Region', 'Ship_mode'])['Shipping_cost'].std()

# Print the results
print("Mean shipping cost per order:")
print(mean_shipping_cost)
print("\nMedian shipping cost per order:")
print(median_shipping_cost)
print("\nStandard deviation of shipping cost per order:")
print(std_shipping_cost)
```

```python

plt.figure(figsize=(12, 7))
sns.barplot(x='Ship_mode', y='Shipping_cost', hue='Region', data=df)
plt.title('Shipping Cost per Order Across Different Modes and Regions')
plt.xlabel('Shipping Mode')
plt.ylabel('Shipping Cost')
plt.legend(title='Region')
plt.tight_layout()
plt.show()
```


#### Shipping costs vary across regions and shipping modes, with notable differences observed, such as higher costs for Same Day shipping in Central Asia and lower costs for Standard Class shipping overall. This highlights the significance of considering both geographical factors and shipping preferences when assessing shipping expenses.
